```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('/content/dataset_enc.csv')
```

```python
df
```

```python
y = OneHotEncoder(sparse_output=False)
y = y.fit_transform(df[['Nama Produk']])
y
```

```python
data_y = pd.DataFrame(y)
data_y
```

```python
header = df['Nama Produk'].sort_values()
data_y.columns = 'Produk_'+ header
data_y
```

```python
data_join = df.join(data_y)
data_join
```
